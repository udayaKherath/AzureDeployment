```python
# get x and y separately
x = df1.drop('Success', axis=1)
y = df1['Success']
```

```python
# split the data set into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
```

```python
#decision tree classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dt = DecisionTreeClassifier()

dt.fit(x_train, y_train)

# Make predictions on the test data
dt_y_pred = dt.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, dt_y_pred)
print("Accuracy:", accuracy)


from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, dt_y_pred))
```

```python
def classify_plan(Marketing_Budget, Customer_Acquisition_Cost, Market_Trend, Sales_Channel,Product_Type,Customer_Base,Competitor,Marketing_Campaign,
                  Financial_Health,Employee_Performance,Investment_Funding,Customer_Feedback):
  market_trend_index = np.where(x.columns==Market_Trend)[0][0]
  Sales_Channel_index = np.where(x.columns==Sales_Channel)[0][0]
  Product_Type_index = np.where(x.columns==Product_Type)[0][0]
  Customer_Base_index = np.where(x.columns==Customer_Base)[0][0]
  Sales_Competitor_index = np.where(x.columns==Competitor)[0][0]
  Marketing_Campaign_index = np.where(x.columns==Marketing_Campaign)[0][0]
  Financial_Health_index = np.where(x.columns==Financial_Health)[0][0]
  Employee_Performance_index = np.where(x.columns==Employee_Performance)[0][0]
  Investment_Funding_index = np.where(x.columns==Investment_Funding)[0][0]
  Customer_Feedback_index = np.where(x.columns==Customer_Feedback)[0][0]

  xt = np.zeros(len(x.columns))
  xt[0] = Marketing_Budget
  xt[1] = Customer_Acquisition_Cost
  if market_trend_index >= 0:
    xt[market_trend_index] = 1

  if market_trend_index >= 0:
    xt[market_trend_index] = 1

  if Sales_Channel_index >= 0:
    xt[Sales_Channel_index] = 1

  if Product_Type_index >= 0:
    xt[Product_Type_index] = 1

  if Customer_Base_index >= 0:
    xt[Customer_Base_index] = 1

  if Sales_Competitor_index >= 0:
    xt[Sales_Competitor_index] = 1

  if Marketing_Campaign_index >= 0:
    xt[Marketing_Campaign_index] = 1

  if Financial_Health_index >= 0:
    xt[Financial_Health_index] = 1

  if Employee_Performance_index >= 0:
    xt[Employee_Performance_index] = 1

  if Investment_Funding_index >= 0:
    xt[Investment_Funding_index] = 1

  if Customer_Feedback_index >= 0:
    xt[Customer_Feedback_index] = 1

  return dt.predict([xt])[0]
```
